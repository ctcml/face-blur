```python
from google_images_download import google_images_download
import os
```

```python
# Constants
DATA_DIR = 'data'
DOWNLOAD_DIR = 'raw'
IMAGE_PATH = os.path.join(os.getcwd(), DATA_DIR, DOWNLOAD_DIR)
CHROMEDRIVER_PATH = '/usr/bin/chromedriver'
```

```python
# Download images from google using google_images_download
response = google_images_download.googleimagesdownload()

arguments = {
    "keywords":"hong kong police brutality, hong kong arrest",
    "limit": 3000,
    "output_directory": DATA_DIR,
    "image_directory": DOWNLOAD_DIR,
    "chromedriver": CHROMEDRIVER_PATH
}
_ = response.download(arguments)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Show an image
img_path = os.path.join(IMAGE_PATH, os.listdir(IMAGE_PATH)[0])
img = mpimg.imread(img_path)
plt.imshow(img)
```

```python
# Change image names for consistency
count = 1
for img_name in os.listdir(IMAGE_PATH):
    img_path = os.path.join(IMAGE_PATH, img_name)
    os.rename(img_path, os.path.join(IMAGE_PATH, '{:0>4}'.format(count)))
    count += 1
```
